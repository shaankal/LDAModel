```python
import numpy as np
import json
import glob
import pandas as pd
import tqdm
import pickle 
import os

#Gensim
import gensim
import gensim.corpora as corpora
from gensim.utils import simple_preprocess
from gensim.models import CoherenceModel
import pyLDAvis.gensim_models as gensimvis

#spacy
import spacy
from nltk.corpus import stopwords
import nltk

#vis
import pyLDAvis
import pyLDAvis.gensim


import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)

nltk.download('stopwords')
stopword = stopwords.words("english")

def load_data(file):
        data = pd.read_csv(file)["body"]
        return data

def lemmatization(texts, allowed_postags=["NOUN", "ADJ", "VERB", "ADV"]):
    nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
    texts_out = []
    for text in texts:
        doc = nlp(text)
        new_text = []
        for token in doc:
            if token.pos_ in allowed_postags:
                new_text.append(token.lemma_)
        final = " ".join(new_text)
        texts_out.append(final)
    return (texts_out)

def gen_words(texts):
    final = []
    for text in texts:
        new = gensim.utils.simple_preprocess(text, deacc=True)
        final.append(new)
    return (final)

# supporting function
def compute_coherence_values(data_words, corpus, id2word, k, a, b):
    
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                        id2word=id2word,
                                        num_topics=k, 
                                        random_state=100,
                                        chunksize=100,
                                        passes=10,
                                        alpha=a,
                                        eta=b)
    print(lda_model.print_topics())
    print("top topics:", lda_model.top_topics(corpus)) #Get the topics with the highest coherence score the coherence for each topic.
    
    """
    topic coherence
    Topic Coherence measures score a single topic by measuring the degree of semantic similarity between high scoring words in the topic. 
    """
    coherence_model_lda = CoherenceModel(model=lda_model, texts=data_words, dictionary=id2word, coherence='c_v')
    #You need to use "data_words" instead of "lemmatized_texts" here, otherwise it will be nan as we saw last time
    
    return coherence_model_lda.get_coherence()

def hyperparam_search(data_words, corpus,id2word):
    grid = {}
    grid['Validation_Set'] = {}

    # Topics range
    min_topics = 1
    max_topics = 3 #11
    step_size = 1
    topics_range = range(min_topics, max_topics, step_size)

    # Alpha parameter
    alpha = list(np.arange(0.01, 1, 0.1))
    alpha.append('symmetric')
    alpha.append('asymmetric')

    # Beta parameter
    beta = list(np.arange(0.01, 1, 0.1))
    beta.append('symmetric')

    # Validation sets
    num_of_docs = len(corpus)
    corpus_sets = [gensim.utils.ClippedCorpus(corpus, int(num_of_docs*0.75)), 
                corpus]

    corpus_title = ['75% Corpus', '100% Corpus']

    model_results = {'Validation_Set': [],
                    'Topics': [],
                    'Alpha': [],
                    'Beta': [],
                    'Coherence': []
                    }

    # Can take a long time to run if you have a lot of data
    if 1 == 1:
        pbar = tqdm.tqdm(total=(len(beta)*len(alpha)*len(topics_range)*len(corpus_title)))
        
        # iterate through validation corpuses
        for i in range(len(corpus_sets)):
            # iterate through number of topics
            for k in topics_range:
                # iterate through alpha values
                for a in alpha:
                    # iterare through beta values
                    for b in beta:
                        # get the coherence score for the given parameters
                        cv = compute_coherence_values(data_words = data_words,corpus=corpus_sets[i], id2word=id2word, 
                                                    k=k, a=a, b=b)
                        # Save the model results
                        model_results['Validation_Set'].append(corpus_title[i])
                        model_results['Topics'].append(k)
                        model_results['Alpha'].append(a)
                        model_results['Beta'].append(b)
                        model_results['Coherence'].append(cv)
                        
                        pbar.update(1)
        os.system("mkdir results")
        pd.DataFrame(model_results).to_csv('results/lda_tuning_results.csv', index=False)
        pbar.close()






```

```python
data = load_data("Lumiere_Shaan/result.csv")
lemmatized_texts = lemmatization(data)
print (lemmatized_texts[0][0:90])
data_words = gen_words(lemmatized_texts)
print (data_words[0][0:20])

id2word = corpora.Dictionary(data_words)

corpus = []
for text in data_words:
    """
    id2word.doc2bow method:
    Convert document (a list of words) into the bag-of-words 
    format = list of (token_id, token_count)
    Each word is assumed to be a tokenized and normalized string.
    No further preprocessing is done on the words in document; 
    Apply tokenization, stemming etc. before calling this method.
    """
    new = id2word.doc2bow(text)
    corpus.append(new)

print ("sample corpus:", corpus[0][0:20])

print ("sample word:",id2word[[0][:1][0]])

hyperparam_search(data_words, corpus, id2word) #you get the best model from search

#assume this is the best model! But you need to write code to evaluate the result from hyperparam search and get the best model
lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                    id2word=id2word,
                                    num_topics=2, 
                                    random_state=100,
                                    chunksize=100,
                                    passes=10,
                                    alpha=0.1,
                                    )
```

```python
def visualize(lda_model, corpus, id2word):
   
    # Visualize the topics. This only works on notebook at the moment...
    pyLDAvis.enable_notebook()
    # you can play with different num of topics as you did in hyperparam search
    num_topics = 1
    LDAvis_data_filepath = os.path.join('results/ldavis_tuned_'+str(num_topics))

    # # this is a bit time consuming - make the if statement True
    # # if you want to execute visualization prep yourself
    LDAvis_prepared = gensimvis.prepare(lda_model, corpus, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)

    # load the pre-prepared pyLDAvis data from disk
    with open(LDAvis_data_filepath, 'rb') as f:
        LDAvis_prepared = pickle.load(f)

    pyLDAvis.save_html(LDAvis_prepared, 'results/ldavis_tuned_'+ str(num_topics) +'.html')

    return LDAvis_prepared
```

```python
LDAvis_prepared=visualize(lda_model, corpus, id2word)
```
